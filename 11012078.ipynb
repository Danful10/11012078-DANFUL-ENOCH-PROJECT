## PLOTS AND GRAPHS OF THE REDWINE DATASET
----------------------

```python
plt.figure(figsize=(10,6))
plt.hist(dataset['alcohol'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Alcohol Content in dataset') 
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()           
```

```python
plt.figure(figsize=(8, 6))
plt.bar(dataset['quality'].value_counts().index, dataset['quality'].value_counts().values, color='salmon', edgecolor='black')
plt.title('Bar Chart of Red Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(dataset['alcohol'], dataset['density'], alpha=0.5, color='green')
plt.title('Scatter Plot of Alcohol Content vs. Density')
plt.xlabel('Alcohol Content')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for quality_value in sorted(dataset['quality'].unique()):
    subset_data = dataset[dataset['quality'] == quality_value]['alcohol']
    plt.boxplot(subset_data, positions=[quality_value], widths=0.6, showfliers=False)

plt.title('Boxplot of Alcohol Content by Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol Content')
plt.xticks(sorted(dataset['quality'].unique()))
plt.show()
```
